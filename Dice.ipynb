```python
def roll_dice(die1, die2, nr_games):
    print('Playing with first die {} and the second die {} for {} games.'.format(die1, die2, nr_games))
    nr_wins = 0
    nr_draws = 0
    nr_losses = 0
    for i in range(nr_games):
        result1 = random.choice(die1)
        result2 = random.choice(die2)
        if result1 > result2:
            nr_wins += 1
        elif result1 == result2:
            nr_draws += 1
        else:
            nr_losses += 1
            
    print('\n\n\nThe first die won {} time and {} draws.'.format(nr_wins, nr_draws))
    return nr_wins, nr_draws, nr_losses

DIE1 = widgets.Select(
    options = ['die 1', 'die 2', 'die 3', 'die 4', 'custom die 1', 'custom die 2'],
    description='first die',
    disabled=False)

DIE2 = widgets.Select(
    options = ['die 1', 'die 2', 'die 3', 'die 4', 'custom die 1', 'custom die 2'],
    description='second die',
    disabled=False)

NrGames = widgets.IntSlider(
    value = 100,
    min = 0,
    max = 10000,
    step = 1,
    description='# games',
    disabled=False)

widgets.Box([DIE1, DIE2, NrGames])
```
